```python
import pandas as pd
import glob
import os
```

```python
# merging the files
files_joined = os.path.join(r'.\Data', "targets*.csv")
```

```python
# Return a list of all joined files
list_files = glob.glob(files_joined)
```

```python
list_files
```

```python
print("** Merging multiple csv files into a single pandas dataframe **")
# Merging files by joining all files
sanction = pd.concat(map(pd.read_csv, list_files), ignore_index=True)
print(sanction)
```

```python
sanction
```

```python
#Saving into a CSV file
sanction.to_csv(r'.\Data\Sanctions_Data_Merged.csv')
```
